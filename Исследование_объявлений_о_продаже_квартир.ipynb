#### Приведём формат вывода даты и времени по столбцу "first_day_exposition"  к удобному для форматирования и обработки

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

#### Произведём подсчёт квартир с нулевым обозначением количества комнат  в столбце "rooms"

```python
data.query('rooms == 0')['rooms'].count()
```

проверим по таблице, к каким данныи относятся 0 комнат

```python
data.query('rooms == 0').head(10)
```

Логично предположить, что с такой площадью и отсутствием площади кухни, 0 комнат означает, что это квартиры-студии.
В стаком случае отсавим всё как есть, здесь ошибок в указании данных нет.

<a id=step_2.3></a>

#### Проверим значения колонки "ceiling_height"

Для начала вычислим среднее значение по высоте потолков ограничив выборку предельно-допустимыми значениями

```python
data.query('ceiling_height > 2 and ceiling_height < 8.3')['ceiling_height'].mean()
```

Приравняем невошедшие значения к среднему, т.к. потолков выше 8,3 м. и ниже 2 м. быть не может

```python
data.loc[data['ceiling_height'] < 2, 'ceiling_height'] = 2.72
```

```python
data.loc[data['ceiling_height'] > 8.3, 'ceiling_height'] = 2.72
```

Проверяем результат:

```python
data.query('ceiling_height < 2 and ceiling_height > 8.3')['ceiling_height'].sum()
```

Отлично, теперь заполним отсутствующие значения в столбце средней величиной, для сохранения статистики по остальным данным в других столбцах

```python
data['ceiling_height'] = data['ceiling_height'].fillna(2.72)
```

Провверим результат

```python
data.info()
```

<a id=step_2.4></a>

#### Выполним проверку столбца "floors_total" на отсутствующие значения

```python
data.loc[data['floors_total'].isna()].head(10)
```

```python
data['floors_total'].isna().sum()
```

Из данной выборки видно только то, что невозможно по остальным праметрам однозначно сказать об этажности дома. По-этому оставим эти данные пока без изменений.

<a id=step_2.5></a>

#### Проверм данные по столбцу "living_area"

```python
data['living_area'].agg(['min','max','mean','median'])
```

Выведем данные для визульного ознаколения, для каких квартир указаны такие значения по жилой площади:

```python
data.query('living_area < 5')
```

Однозначно определить какое значение следует указать мы не можем, но ясно одно, что остальной массив даннх лучше сохранить, по-этому заменим неправдоподобные значения на None

```python
data.loc[data['living_area'] < 5, 'living_area'] = None
```

```python
data['living_area'].isna().sum()
```

проверим результат применения None к выбранным строкам:

```python
data.query('living_area < 5')
```

Отлично, неправдоподобные данные по площади жилого помещения устранены. К сожалению таких пропущенных значений в столбце много, но удалять строки с другими данными вредно для статистики, оставим строки с неизвестными площадями как есть.

<a id=step_2.6></a>

#### Исследуем столбец "kitchen_area" на предмет корректных значений 

Согласно Снип, кухня не может быть менее 5 м2. Выполним выборку по данному значению:

```python
data.query('kitchen_area < 5')
```

```python
data['kitchen_area'].isna().sum()
```

Заполним пропуски в значениях нулевыми учитывая, что отсутствие данных о площади кухни связано с квартирами, где комнаты объеденины с кухней и площадь комнаты уже учитывает площадь кухни.

```python
data['kitchen_area'] = data['kitchen_area'].fillna(0)
```

Вычислим среднее значение для кухни отсортировав "мусорные" значения

```python
MEAN_AREA = data.query('kitchen_area > 5' and 'kitchen_area < 15')['kitchen_area'].mean()
```

Заменим ошибочные данные средним значением нашей выборки

```python
data.loc[data['kitchen_area'] < 5, 'kitchen_area'] = MEAN_AREA
```

Проверяем результат:

```python
data.query('kitchen_area < 5')
```

#### Исследуем колонку "balcony" на предмет отсутствующих значений

```python
data['balcony'].isna().sum()
```

Ого, пропусков около 50% от всех данных, это говорит о том, что такие пропуски не могут быть случайны, ознакомимся со смежными данными по отсутствующим значениям колонки "balcony"

```python
data.loc[data['balcony'].isna()]
```

Судя по распределению данных, логично было бы предположить, что те данные которые имеют пропуск - это означает отсутствие балкона, таким образом можем указать 0 вместо NaN
